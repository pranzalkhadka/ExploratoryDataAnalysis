```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("Iris.csv")
```

```python
iris.shape
```

```python
iris.columns
```

```python
iris["Species"].value_counts()
```

```python
iris.plot(kind="scatter",x="SepalLengthCm",y="SepalWidthCm")
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris,hue="Species").map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend()
plt.show()
```

We can see that using just sepal length and sepal width we can distinguish iris setosa and iris(versicolor,virginica)

```python
iris.plot(kind="scatter",x="PetalLengthCm",y="PetalWidthCm")
```

```python
sns.FacetGrid(iris,hue="Species").map(plt.scatter,"PetalLengthCm","PetalWidthCm").add_legend()
```

Using petal length and petal width we can clearly distinguish all three species of iris flowers easily . SO petal length and petal width looks like a more important feature in this case 

Visualizing 3d or 4d plots is very complex . So we instead use pair plots

```python
sns.pairplot(iris,hue="Species",height=3)
```

A limitation of pair plot is that if the number of features increases visualizing these plots can be extremely difficult

```python
sns.FacetGrid(iris,hue="Species",height=5).map(sns.histplot,"PetalLengthCm").add_legend()
```

We can clearly separate setosa here but there is some intersection between pdf of virginica and versicolor . A value of 4.8 can act as a good threshhold to separate virginica and versicolor

```python
sns.FacetGrid(iris,hue="Species",height=5).map(sns.histplot,"PetalWidthCm").add_legend()
```

Petal width also can separate iris flowers quite good . So it also looks like a important feature

```python
sns.FacetGrid(iris,hue="Species",height=5).map(sns.histplot,"SepalLengthCm").add_legend()
```

```python
sns.FacetGrid(iris,hue="Species",height=5).map(sns.histplot,"SepalWidthCm").add_legend()
```

The pdf of all iris species are highly intersecting if we take sepal length or sepal width . so these features looks less important than petal length and petal width
